```python
# importing libraies for analysiing and processing data

import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/Data Science Tutedude/austin_weather_clean.csv')
data.head()
```

```python
x = data.drop(['PrecipitationSumInches'], axis=1)
```

```python
y = data['PrecipitationSumInches']
```

```python
# Reshaping it into 2d vector
# x = x.values.reshape(-1, 1)
y = y.values.reshape(-1, 1)
```

```python
y
```

```python
# random days for observation
days_index = 798
days = [i for i in range(y.size)]
```

```python
# Initialise the linear regression classifier
regressor = LinearRegression()

# train the classifier
regressor.fit(x, y)
```

```python
# Plot a graph
print("The precipitation trend graph :")
plt.scatter(days, y, color='red')
plt.scatter(days[days_index], y[days_index], color='green')
plt.title('Precipitation Trend')
plt.xlabel('Days')
plt.ylabel('Precipitation')

plt.show()
x_vis = x.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
# plt.plot(days, regressor.predict(x_vis), color='blue')
# plt.plot(days, regressor.predict(x), color='blue')
```

```python
print("The perception vs attributes trend graph: ")
plt.figure(figsize=(12, 8))
for i in range(x_vis.columns.size):
    plt.subplot(3, 2, i + 1)
    plt.scatter(days, x_vis[x_vis.columns.values[i]], color='g')
    plt.scatter(days[days_index], x_vis.iloc[days_index, i], color='r')
    plt.title(x_vis.columns.values[i])
    plt.xlabel('Days')
    plt.ylabel(x_vis.columns.values[i])

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
