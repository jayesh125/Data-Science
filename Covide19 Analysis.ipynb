# **Covid19 Data Analysis**

```python
# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the datasets
dataset = pd.read_csv('/content/drive/MyDrive/Data Science Tutedude/covid19_Confirmed_dataset.csv')
dataset.head()
```

```python
dataset.shape
```

```python
# delete the useless columns

df = dataset.drop(['Lat','Long'],axis=1)
df.head()
```

```python
# aggrigate the rows by country

corona_dataset_aggregated = dataset.drop(columns=["Province/State", "Lat", "Long"]).groupby("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
# Aggregate the rows by the country
corona_dataset_aggregated.loc['China'].plot(label='China')
corona_dataset_aggregated.loc['India'].plot(label='India')
corona_dataset_aggregated.loc['Spain'].plot(label='Spain')

plt.legend()
plt.show()
```

```python
# Calculate a good measure
corona_dataset_aggregated.loc['China'].plot()
```

```python
# Corona spread in china for first 3 days
corona_dataset_aggregated.loc['China'][:3].plot()
```

```python
# Calculate the first derivative of the curve

corona_dataset_aggregated.loc['China'].diff().plot()
```

```python
# maximum infection rate
max_infection_china = corona_dataset_aggregated.loc['China'].diff().max()
max_infection_india = corona_dataset_aggregated.loc['India'].diff().max()
max_infection_spain = corona_dataset_aggregated.loc['Spain'].diff().max()

print(f"Max Infection Rate for China : {max_infection_china}")
print(f"Max Infection Rate for India : {max_infection_india}")
print(f"Max Infection Rate for Spain : {max_infection_spain}")
```

```python
# maximum infection rate for all countries and storing it in dataset
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for country in countries:
  max_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())

corona_dataset_aggregated['max_infection_rates'] = max_infection_rates

corona_dataset_aggregated
```

```python
# Create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated['max_infection_rates'])
corona_data
```

```python
# happines report datasets
happiness_report = pd.read_csv('/content/drive/MyDrive/Data Science Tutedude/worldwide_happiness_report.csv')
happiness_report
```

```python
# Drop useless columns
useless_cols = ['Overall rank', 'Score', 'Generosity', 'Perceptions of corruption']
```

```python
happiness_report.drop(useless_cols, axis=1, inplace=True)
happiness_report
```

```python
happiness_report.set_index('Country or region', inplace=True)
happiness_report.head()
```

```python
# join the datsets
corona_data.shape
```

```python
happiness_report.shape
```

```python
data = corona_data.join(happiness_report, how='inner')
data.head()
```

```python
# Correlation between this factors

data.corr()
```

```python
# Plotting the Graph between GDP and maximum infection rate
x = data['GDP per capita']
y = data['max_infection_rates']

# Plotting the scatter plot with log-transformed infection rates
sns.scatterplot(x=x, y=np.log(y))

# Adding labels and title
plt.xlabel('GDP per capita')
plt.ylabel('Log(Max Infection Rates)')
plt.title('GDP per capita vs Max Infection Rates')

# Show the plot
plt.show()
```

```python
sns.regplot(x=x, y=np.log(y))
```
